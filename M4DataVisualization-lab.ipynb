# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt

# Create a cursor object to execute SQL queries
cursor = conn.cursor()

# Query the database to get data you want to plot (e.g., 'ConvertedComp' column)
query = "SELECT ConvertedComp FROM master WHERE ConvertedComp IS NOT NULL;"
data = cursor.execute(query).fetchall()

# Extract the values from the result set
values = [row[0] for row in data]

# Create a histogram
plt.hist(values, bins=20, edgecolor='k')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.title('Histogram of Converted Compensation')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

# Query the database to get the 'Age' column data
query = "SELECT Age FROM master WHERE Age IS NOT NULL;"
data = cursor.execute(query).fetchall()

# Extract the values from the result set
values = [row[0] for row in data]

# Create a box plot
plt.boxplot(values, vert=False)
plt.xlabel('Age')
plt.title('Box Plot of Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

# Query the database to get the 'Age' and 'WorkWeekHrs' columns data
query = "SELECT Age, WorkWeekHrs FROM master WHERE Age IS NOT NULL AND WorkWeekHrs IS NOT NULL;"
data = cursor.execute(query).fetchall()

# Extract the values from the result set
ages = [row[0] for row in data]
work_hours = [row[1] for row in data]

# Create a scatter plot
plt.scatter(ages, work_hours, alpha=0.5)
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.title('Scatter Plot of Age vs. WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

# Query the database to get the 'Age', 'WorkWeekHrs', and 'CodeRevHrs' columns data
query = "SELECT Age, WorkWeekHrs, CodeRevHrs FROM master WHERE Age IS NOT NULL AND WorkWeekHrs IS NOT NULL AND CodeRevHrs IS NOT NULL;"
data = cursor.execute(query).fetchall()

# Extract the values from the result set
ages = [row[0] for row in data]
work_hours = [row[1] for row in data]
code_review_hours = [row[2] for row in data]

# Calculate bubble sizes based on the 'Age' column
bubble_sizes = [age * 2 for age in ages]  # You can adjust the multiplier for the bubble size

# Create a bubble plot
plt.scatter(work_hours, code_review_hours, s=bubble_sizes, alpha=0.5)
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.title('Bubble Plot of WorkWeekHrs vs. CodeRevHrs (Age as Bubble Size)')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

# Query the database to get the top 5 databases respondents wish to learn next year
query = """
SELECT DatabaseDesireNextYear, COUNT(*) as count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5;
"""

data = cursor.execute(query).fetchall()

# Extract the database names and counts from the result set
database_names = [row[0] for row in data]
counts = [row[1] for row in data]

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=database_names, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the pie chart
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

# Query the database to get the WorkWeekHrs and CodeRevHrs data for the age group 30 to 35
query = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age >= 30 AND Age <= 35;
"""

data = cursor.execute(query).fetchall()

# Extract the data into a DataFrame
df = pd.DataFrame(data, columns=['Age', 'WorkWeekHrs', 'CodeRevHrs'])

# Calculate the median values for WorkWeekHrs and CodeRevHrs
median_work_week_hrs = df['WorkWeekHrs'].median()
median_code_rev_hrs = df['CodeRevHrs'].median()

# Create a stacked chart
plt.figure(figsize=(10, 6))

# Stacked bar chart with median values
plt.bar('30-35', median_work_week_hrs, label='Median WorkWeekHrs')
plt.bar('30-35', median_code_rev_hrs, bottom=median_work_week_hrs, label='Median CodeRevHrs')

plt.xlabel('Age Group')
plt.ylabel('Median Hours')
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.legend()

# Show the stacked chart
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

# Query the database to get the data for ages from 45 to 60
query = """
SELECT Age, ConvertedComp
FROM master
WHERE Age >= 45 AND Age <= 60;
"""

data = cursor.execute(query).fetchall()

# Extract the data into a DataFrame
df = pd.DataFrame(data, columns=['Age', 'ConvertedComp'])

# Calculate the median ConvertedComp for each age group
median_values = []
age_groups = list(range(45, 61))

for age in age_groups:
    median_value = df[df['Age'] == age]['ConvertedComp'].median()
    median_values.append(median_value)

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(age_groups, median_values, marker='o', linestyle='-')

plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp for Ages 45 to 60 (Manually Calculated)')
plt.grid(True)

# Show the line plot
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

# Query the database to get the counts of each unique value in the MainBranch column
query = """
SELECT MainBranch, COUNT(*) as Count
FROM master
GROUP BY MainBranch
ORDER BY Count DESC;
"""

data = cursor.execute(query).fetchall()

# Extract the data into a DataFrame
df = pd.DataFrame(data, columns=['MainBranch', 'Count'])

# Create a horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(df['MainBranch'], df['Count'], color='skyblue')

plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.title('Count of Respondents by MainBranch')

# Show the horizontal bar chart
plt.show()
```
